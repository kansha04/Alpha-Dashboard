# The Challenge: "The Broken Ticker"
You are receiving a price feed for a stock, but the data is "dirty." If you try to do math with it right now, your program will crash.

### The Data:

Python
raw_price = "  $ 1,299.50 (USD)  "
Your Mission: Write a function called clean_price that takes this string and returns the number 1299.5 (as a float, not a string).

Constraints:
Get rid of the whitespace.
Get rid of the "$" and "(USD)".
Tricky Part: Get rid of the comma ,. (Python cannot convert "1,299.50" to a number directly; it hates commas).
Convert the final clean string into a float.

```python
raw_price = "  $ 1,299.50 (USD)  "

# Function called clean_price that takes this string and returns the number 1299.5 (as a float, not a string).
def clean_price(raw_price):
    # Get rid of the "$" and "(USD)"
    clean_price = raw_price.replace("$", "")
    clean_price = clean_price.replace("(USD)", "")
    # Get rid of whitespace
    clean_price = clean_price.strip()
    # Get rid of the comma.
    clean_price = clean_price.replace(",", "")
    
    cleaned_price = float(clean_price)
    
    return cleaned_price
    
print(clean_price(raw_price))
```
